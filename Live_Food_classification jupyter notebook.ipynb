```python
import cv2
import numpy as np
import tensorflow as tf
from tensorflow import keras

model = keras.models.load_model("D:\Food_classification_model.h5")

class_labels = ['Bread', 'Dairy product', 'Dessert', 'Egg', 'Fried food', 'Meat', 'Noodles-Pasta', 'Rice', 'Seafood', 'Soup', 'Vegetable-Fruit']

def preprocess_image(image):
    image = cv2.resize(image, (224, 224))

    image = image.astype(np.float32)

    image /= 255.0
    image = np.expand_dims(image, axis=0)
    return image


def classify_image(image):
    image = cv2.resize(image, (224, 224))  
    image = image.astype('float32') / 255.0  
    image = np.expand_dims(image, axis=0)  

    predictions = model.predict(image)
    return predictions

```

```python
cap = cv2.VideoCapture(0)
```

```python
while True:
    ret, frame = cap.read()  # Capture a frame
    predictions = classify_image(frame)  # Classify the frame
    
    predicted_class_index = np.argmax(predictions)
    predicted_class_name = class_labels[predicted_class_index]
    
    label = f"Class: {predicted_class_name}"
    cv2.putText(frame, label, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    cv2.imshow('Live Classification', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):  # Press 'q' to exit
        break
```

```python
cap.release()
cv2.destroyAllWindows()
```
